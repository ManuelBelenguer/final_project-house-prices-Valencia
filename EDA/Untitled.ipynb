```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")
```

```python
data = pd.read_csv('../modeling/data_ready.csv')
```

```python
corr = data.corr('pearson')

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

pearson_meth =data.corr('pearson')
fig, ax = plt.subplots(figsize=(8,8))
ax = sns.heatmap(corr, mask=mask, annot=True, fmt='.2%')
plt.show()
```

##### All the variables (apart from longitude) present some kind of correlation with the dependent variable.

### Distribution plots

```python
continous = data[['sqft_surface', 'bedrooms', 'bathrooms', 'price']]
```

```python
for x in continous:
    print(x)
    sns.distplot(continous[x])
    plt.show()
```

### Outliers

```python
for col in continous[['sqft_surface', 'price']]:
        sns.boxplot(continous[col])
        plt.show()
```
